```python
import pandas as pd
import altair as alt
import pathlib
import os
```

```python
# Retrieve data
current_dir = pathlib.Path(os.getcwd())
data_dir = current_dir.parents[0]/"data"
filename = data_dir / "clean_dataset.csv"

poverty_df = pd.read_csv(filename)
```

```python
# Create horizontal bar plots
def horizontal_bars_faceted(df, variable_of_interest):
    source = df
    variable_str_clean = str(variable_of_interest).replace("_", " ").capitalize()
    title_label = variable_str_clean + " by State and Year"
    chart = alt.Chart(source).mark_bar().encode(
        x=alt.X(variable_of_interest, title=title_label),
        y=alt.Y("state_name", title="State"),
        color=alt.Color('extreme_poverty:Q', scale=alt.Scale(domain=[0, 0.5], range=['green', 'red']))
    )
    faceted_chart = chart.facet(column=alt.Column("year:O", title="Year")).properties(title=variable_str_clean + " by State and Year")
    return faceted_chart

horizontal_bars_faceted(poverty_df, "extreme_poverty")
```
